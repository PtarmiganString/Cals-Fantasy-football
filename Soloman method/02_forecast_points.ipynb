```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import  r2_score
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 100)
```

# Import data

```python
historic_df = pd.read_csv('Data/historic.csv').dropna(subset = ['name','position','FIX_Team','FIX_Opponent','FIX_Home?', 'FIX_GW', 'FIX_season','minutes','total_points'])

historic_df = historic_df.drop_duplicates(subset = ['name','position','FIX_Team','FIX_Opponent','FIX_Home?', 'FIX_GW', 'FIX_season','minutes','total_points'])

historic_df['FIX_season'].unique()
```

# Prepare dataset
- Player
- Position
- Team
- Opposition
- Home/away
- Gameweek
- Season
- Minutes

```python
def encode(df):

    for col in df.select_dtypes(exclude = 'number').columns:

        print("Encoding ", col)

        lb_make = LabelEncoder()
        df[col +'_encoded'] = lb_make.fit_transform(df[col].astype(str))
        
    return df
```

```python
historic_df = encode(historic_df)
```

```python
train_df, test_df = train_test_split(historic_df, test_size=0.33, random_state=42)
```

```python
X_train = train_df[['name_encoded','position','FIX_Team_encoded','FIX_Opponent_encoded','FIX_Home?', 'FIX_GW', 'FIX_season_encoded','minutes']]
y_train = train_df['total_points']
```

```python
X_test = test_df[['name_encoded','position','FIX_Team_encoded','FIX_Opponent_encoded','FIX_Home?', 'FIX_GW', 'FIX_season_encoded','minutes']]
y_test = test_df['total_points']
```

# Evaluate

```python
def evaluate(eval_df, seasons = None):
    
    if seasons is not None:
        eval_df = eval_df.loc[eval_df['FIX_season'].isin(seasons)]

    print("Seasons included :", eval_df['FIX_season'].unique())
    
    eval_df = eval_df.sort_values(['name','FIX_Fixture_date'])

    eval_df = eval_df.groupby(['name','FIX_season'])[['total_points','prediction']].sum().reset_index()
        
    eval_df['rank_actual'] = eval_df['total_points'].rank(ascending = False, method = 'min')
    eval_df['rank_prediction'] = eval_df['prediction'].rank(ascending = False, method = 'min')

    plot_results(eval_df['total_points'],eval_df['prediction'])
    
    return eval_df.sort_values('total_points', ascending = False).head()
    
```
